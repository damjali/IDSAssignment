```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes=plt.subplots(4,2, figsize=(16,16))
sns.countplot(x='Gender',data=df,ax=axes[0,0])
sns.countplot(x='Married',data=df,ax=axes[0,1])
sns.countplot(x='Dependents',data=df,ax=axes[1,0])
sns.countplot(x='Education',data=df,ax=axes[1,1])
sns.countplot(x='Self_Employed',data=df,ax=axes[2,0])
sns.countplot(x='Property_Area',data=df,ax=axes[2,1])
sns.countplot(x='Loan_Status',data=df,ax=axes[3,0])

fig.delaxes(axes[3,1])
```

```python
fig, axes=plt.subplots(3,2, figsize=(16,16))

sns.histplot(x='ApplicantIncome',data=df,ax=axes[0,0])
sns.histplot(x='CoapplicantIncome',data=df,ax=axes[0,1])
sns.histplot(x='LoanAmount',data=df,ax=axes[1,0])
sns.histplot(x='Loan_Amount_Term',data=df,ax=axes[1,1])
sns.histplot(x='Credit_History',data=df,ax=axes[2,0])

fig.delaxes(axes[2,1])
```

```python
fig, axes=plt.subplots(3,2, figsize=(16,16))

sns.boxplot(x='ApplicantIncome',data=df,ax=axes[0,0])
sns.boxplot(x='CoapplicantIncome',data=df,ax=axes[0,1])
sns.boxplot(x='LoanAmount',data=df,ax=axes[1,0])
sns.boxplot(x='Loan_Amount_Term',data=df,ax=axes[1,1])
sns.boxplot(x='Credit_History',data=df,ax=axes[2,0])

fig.delaxes(axes[2,1])
```
